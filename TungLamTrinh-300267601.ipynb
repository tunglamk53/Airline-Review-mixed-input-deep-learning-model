# PART 2: Baseline Model 1 Using Numeric and Categorical Features

```python
# Combine scaled numeric features and dummy categorical features into a dataframe

# combining with reset index
combined_df = pd.concat([tran_num_df.reset_index(drop=True), 
                         dum_cat_df.reset_index(drop=True)], 
                        axis=1)

# add dummy out-put 'recommended' to combined dataframe
combined_df['recommended'] = dum_recommended_df['recommended'].reset_index(drop=True)

# combined_df
```

```python
combined_df.head(3)
```

```python
# Split the data into 75% training sample and 25% test sample
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(combined_df,
                                     test_size = 0.25,
                                     random_state = 0)

print('df_train: ', df_train.shape)
print('df_test: ', df_test.shape)
```

```python
# Logististic Regression, Nearest Neighbors, Linear SVM, MLPClassifier, Decision Tree, Random Forest, and XGBoost

!pip install xgboost
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
```

```python
# in-put dataframe "df_train" & "df_test"
X_train = df_train.drop(['recommended'], axis=1)
x_test = df_test.drop(['recommended'], axis=1)

# out-put dataframe
Y_train = df_train['recommended']
y_test = df_test['recommended']
```

```python
# Model Pipeline
```

```python
names = ["Logististic Regression", 
         "Nearest Neighbors", 
         "Linear SVM", 
         "RBF SVM", 
         "Decision Tree", 
         "Random Forest", 
         "XGBoost"]
         
scores = []

classifiers = [
        LogisticRegression(),
        KNeighborsClassifier(3),
        SVC(kernel="linear", C=0.025),
        SVC(gamma=2, C=1),
        DecisionTreeClassifier(max_depth=5),
        RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
        XGBClassifier(),
]

for classifier in classifiers:
    pipeline = Pipeline(
                    steps=[('classifier', classifier)]
                )
    pipeline.fit(X_train, Y_train)
    print(classifier)
    scores.append(pipeline.score(x_test, y_test))
    print("model score: %.3f" % pipeline.score(x_test, y_test))
    print("\n -----------------------------------------------------------------------------------")
```

```python
scores_df = pd.DataFrame(
                zip(names, scores), 
                columns=['Classifier', 'Accuracy Score']
            )
```

# PART 4: Mixed Input Deep Learning Model

```python
# combination of scaled numerical data and dummy categorical data
print(combined_df.shape)
# TF-IDF Vectorizer text data
print(tfidf_text_df.shape)
# output feature 'Recommended' YES/NO
print(dum_recommended_df.shape)
```

```python
# split the data into training and test samples
x_train_ni, x_test_ni, x_train_ti, x_test_ti, y_train_recommended, y_test_recommended = train_test_split(
                                                    np.array(combined_df), 
                                                    np.array(tfidf_text_df), 
                                                    np.array(dum_recommended_df['recommended']), 
                                                    test_size=0.25,random_state=42)
```

```python
#Install keras
#!pip install keras
#keras deep learning
import keras
from keras.models import Sequential, Model
from keras.layers import LSTM, Activation, Dense, Dropout, Input, Embedding, Flatten
from keras.layers import Bidirectional, GlobalMaxPooling1D, concatenate
from keras.optimizers import RMSprop, Adam
from keras.callbacks import EarlyStopping
from keras.losses import SparseCategoricalCrossentropy
from keras.preprocessing import sequence
from keras import regularizers
```

```python
# MLP INPUT
mlp_input = Input(shape= (x_train_ni.shape[1], ))

# Hidden layers (3 layers, each with 100 units and "relu" activation function)
mlp_layer1 = Dense(100, activation="relu")(mlp_input)
mlp_layer2 = Dense(100, activation="relu")(mlp_layer1)
mlp_layer3 = Dense(100, activation="relu")(mlp_layer2)
# mlp_output
mlp_out = Dense(100, activation="relu")(mlp_layer3)

# LSTM Input (text features)
lstm_input = Input(shape= (x_train_ti.shape[1], 1))

# hidden layer
lstm_out = LSTM(64,
                dropout=0.3,
                recurrent_dropout=0.3,
                kernel_regularizer=regularizers.l2(0.01))(lstm_input)

# concatenate the outputs from the 2 branches
combined_input = concatenate([mlp_out, lstm_out])

# define the fully connected layer
fc = Dense(4, activation='relu')(combined_input)

# add a classifier (multiclass == "softmax"), unit 1 as it's the binary output (Recommended='YES/NO')
cl = Dense(1, activation="softmax")(fc)

# define model
model = Model(inputs=[mlp_input, lstm_input], outputs=cl)


# compile the model
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# print model summary
print(model.summary())
```

```python
# training the model
# fit the model using batch_size of 64 and 2 epochs

# x_train_ni1 = x_train_ni.reshape(x_train_ni.shape[0], 1, x_train_ni.shape[1])
# x_train_ti1 = x_train_ti.reshape(x_train_ti.shape[0], 1, x_train_ti.shape[1])
# x_train_ni1 = x_train_ni.reshape(x_train_ni.shape[1], )
# x_train_ti1 = x_train_ti.reshape(x_train_ti.shape[1], 1)

history = model.fit(x = [x_train_ni1, x_train_ti1],
                    y = y_train_recommended,
                    validation_data=([x_test_ni, x_test_ti], y_test_recommended),
                    epochs=2,
                    batch_size=64)
```

```python
#compute overall model loss and accuracy scores
accr = model.evaluate([x_test_ni, x_test_ti], y_test_recommended)
print('Test Set\n  Loss: {:0.3f}\n   Accuracy:{:0.3f}'.format(accr[0], accr[1]))
```
